1. To what does a relative path refer?

A relative path specifies the location of a file or directory with respect to the current location of the user or program, rather than with respect to the root directory of the file system.

2. What does an absolute path start with your operating system?

An absolute path starts with the root directory of the file system such as / or C: depending upon the system windows(C:) or mac(/)

3. What do the functions os.getcwd() and os.chdir() do?

These functions are used to get and change the current working directory in a Python program.

4. What are the . and .. folders?

. folders refer to current directory while .. folders refer to parent directory

5. In C:\bacon\eggs\spam.txt, which part is the dir name, and which part is the base name?

The directory name in this file path is bacon\eggs, and the base name is spam.txt. 

6. What are the three “mode” arguments that can be passed to the open() function?

Three mode arguments that can be passed to the open() function are r(read), w(write),a(append)

7. What happens if an existing file is opened in write mode?

The entire content of the file will be deleted and replaced with the new content that is written to the file.

8. How do you tell the difference between read() and readlines()?

The read() method returns the contents of the file as a single string, while the readlines() method returns the contents of the file as a list of strings, where each string represents a single line from the file.

9. What data structure does a shelf value resemble?

A shelf value in Python resembles a dictionary data structure but with the added feature of persistent storage on disk.
